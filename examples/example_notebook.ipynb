```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import matplotlib.ticker as ticker
import sympy as sp
from scipy.special import jv
from scipy.optimize import fsolve

plt.style.use(['science','notebook', 'grid'])
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['figure.figsize'] = [7, 4]
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.family'] = 'Roboto'
plt.rcParams['text.usetex'] = True
plt.rcParams['xtick.minor.visible'] = False
plt.rcParams['ytick.minor.visible'] = False
```

```python
dat = pd.read_csv(f'./example_P01_fidYpi2_NK20.dat', delimiter='\t', index_col=False, skiprows=1)
A = dat['#A/w_q'].unique()
w = dat['w/w_q'].unique()
P01 = dat['P01_T'].values.reshape(len(A), len(w)).T

plt.contourf(A, w, P01, levels=200, cmap='seismic', vmin=0, vmax=1)
plt.xlabel('$A/\\omega_q$')
plt.ylabel('$\\omega/\\omega_q$')
plt.colorbar(label='$P_{01}$')
plt.show()

F = dat['Fidelidad'].values.reshape(len(A), len(w)).T
Err = 1 - F
plt.contourf(A, w, np.log10(Err), levels=200, cmap='turbo_r')
plt.xlabel('$A/\\omega_q$')
plt.ylabel('$\\omega/\\omega_q$')
plt.colorbar(label='$\\log_{10}(1 - F)$')
plt.show()
```
